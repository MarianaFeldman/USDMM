```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Versão a ser lida
v = 'V3'
```

# Gráfico erro nas épocas do reticulado das janelas

```python
error_ep_w = np.load('error_ep_w_'+v+'.txt', allow_pickle=True)

coluna_x = 'epoch'
coluna_y1 = 'error_train'
coluna_y2 = 'error_val'

plt.figure(figsize=(10, 6)) 
plt.plot(error_ep_w[coluna_x], error_ep_w[coluna_y1], label='error_train', marker='o')
plt.plot(error_ep_w[coluna_x], error_ep_w[coluna_y2], label='error_val', marker='s')

plt.xlabel('Época')
plt.ylabel('Erro')
plt.title('Erro Épocas reticulado das janelas')

plt.legend()

plt.grid(True)  # Adicione grades se desejar
plt.show()
```

# Gráfico erro nas épocas do reticulado das funções booleanas

## Janela Especifica

```python
janela = 1
```

```python
error_ep_f = np.load('error_ep_f_'+v+'.txt', allow_pickle=True)
df_error_ep_f = pd.DataFrame(error_ep_f)

coluna_x = 'epoch'
coluna_y = 'error'


plt.figure(figsize=(10, 6)) 
plt.plot(df_error_ep_f[df_error_ep_f.W_key==janela][coluna_x], df_error_ep_f[df_error_ep_f.W_key==1][coluna_y], label='error_train', marker='o')

plt.xlabel('Época')
plt.ylabel('Erro')
plt.title('Erro Épocas reticulado das funções')

plt.legend()

plt.grid(True)  # Adicione grades se desejar
plt.show()
```

## Todas as Janelas

```python
error_ep_f = np.load('error_ep_f_'+v+'.txt', allow_pickle=True)
df_error_ep_f = pd.DataFrame(error_ep_f)

for w in df_error_ep_f.W_key.unique():
    print(f"Janela {w}")

    coluna_x = 'epoch'
    coluna_y = 'error'


    plt.figure(figsize=(10, 6)) 
    plt.plot(df_error_ep_f[df_error_ep_f.W_key==w][coluna_x], df_error_ep_f[df_error_ep_f.W_key==w][coluna_y], label='error_train', marker='o')

    plt.xlabel('Época')
    plt.ylabel('Erro')
    plt.title('Erro Épocas reticulado das funções')

    plt.legend()

    plt.grid(True)  # Adicione grades se desejar
    plt.show()
    print('----')
```

# Gráfico de todas as Janelas visitadas

```python
w_hist = np.load('W_hist_'+v+'.txt', allow_pickle=True)
df = pd.DataFrame(w_hist)
```

```python
df.f_epoch_min.value_counts()
```

```python
df[['error_train', 'error_val']] = df['error'].apply(pd.Series)
df.drop(columns=['error'], inplace=True)
df.head()
```

## Erros nas Janelas

```python
coluna_x = 'W_key'
coluna_y1 = 'error_train'
coluna_y2 = 'error_val'

# Criar o gráfico de linhas
plt.figure(figsize=(10, 6))  # Ajuste o tamanho da figura conforme necessário
plt.plot(df[coluna_x], df[coluna_y1], label='error_train', marker='o')
plt.plot(df[coluna_x], df[coluna_y2], label='error_val', marker='s')

# Adicionar rótulos e título
plt.xlabel('Janela_visitada')
plt.ylabel('Erro')
plt.title('Erro das janelas visitadas')

# Adicionar legenda
plt.legend()

# Mostrar o gráfico
plt.grid(True)  # Adicione grades se desejar
plt.show()
```
